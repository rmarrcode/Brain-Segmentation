```python
def encoder_block(inp, max_pool, in_channels):
    block = [
        torch.nn.Conv2d(in_channels=in_channels, out_channels=64, kernel_size=3),
        torch.nn.ReLU(),
        torch.nn.Conv2d(in_channels=64, out_channels=64, kernel_size=3),
        torch.nn.ReLU()]
    if max_pool:
        block.append(torch.nn.MaxPool2d(2,2))
    return torch.nn.Sequential(*block)


```

```python
test_load = nib.load('/Users/ryanmarr/Downloads/NFBS_Dataset/A00052340/sub-A00052340_ses-NFB3_T1w.nii.gz').get_fdata()
tens = torch.from_numpy(test_load)
out = encoder_block(tens, True, 1)
print(out)
```
